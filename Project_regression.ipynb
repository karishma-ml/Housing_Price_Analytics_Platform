```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("MedHouseVal" , axis = 1)
```

```python
y = df.MedHouseVal
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size = 0.2 , random_state = 4)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_x_train = scaler.fit_transform(x_train)
scaler_x_test = scaler.transform(x_test)
```

```python
scaler_x_train[:2]
```

```python
scaler_x_test[:2]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() 
```

```python
lr.fit(scaler_x_train , y_train)
```

```python
y_pred = lr.predict(scaler_x_test)
```

```python
y_pred
```

```python
# regression model evaluation metrics 
from sklearn.metrics import r2_score , mean_absolute_error ,mean_squared_error
```

```python
r2_lr = round(r2_score(y_test , y_pred) , 2)
```

```python
r2_lr
```

```python
#mean_absolute_error
mean_lr =round(mean_absolute_error(y_test , y_pred),2)
```

```python
mean_lr
```

```python
#mean_squared_error
mse_lr=round(mean_squared_error(y_test , y_pred) , 3)
```

```python
mse_lr
```

```python
#root mean_squared_error
rmse_lr = round(np.sqrt(mean_squared_error(y_test , y_pred)),2)
```

```python
print(rmse_lr)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
ridge
```

```python
y_pred = ridge.predict(scaler_x_test)
```

```python
y_pred
```

```python
ridge_r2  = round (r2_score(y_pred , y_test ) ,2)
```

```python
ridge_r2
```

```python
mse
```

```python
mae
```
